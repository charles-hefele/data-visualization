# t-SNE Data Exploration Notebook
---
##### Adopted from https://www.datatechnotes.com/2020/11/tsne-visualization-example-in-python.html

```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_iris
from sklearn.datasets import load_digits
from sklearn.datasets import load_wine
from sklearn.datasets import load_breast_cancer
from keras.datasets import mnist
from keras.datasets import fashion_mnist
from numpy import reshape
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set(rc = {'figure.figsize':(15,8)})
```

# 1. Iris dataset (150 samples, 4 features, 3 classes)
### Runtime: ~1s

```python
# load iris dataset
iris = load_iris()
x = iris.data
y = iris.target

# run t-SNE
tsne = TSNE(n_components=2, verbose=0)
%time z = tsne.fit_transform(x)

# load to a dataframe
df = pd.DataFrame()
df["y"] = y
df["comp-1"] = z[:,0]
df["comp-2"] = z[:,1]

# plot it
fig, ax = plt.subplots()
sns.scatterplot(x="comp-1", y="comp-2", hue=df.y.tolist(),
                palette=sns.color_palette("hls", 3),
                data=df).set(title="Iris data t-SNE projection")

# change to iris labels
for t, l in zip(ax.legend().texts, iris.target_names):
    t.set_text(l)
```

# 5. MNIST limited dataset (3000 samples, 784 features, 10 classes)
### Runtime: ~38s

```python
# load first 3000 rows of the MNIST training set
(x_train, y_train), (_, _) = mnist.load_data()
x_train = x_train[:3000]
y_train = y_train[:3000]
print(x_train.shape)

# reshape it
x_mnist = reshape(x_train, [x_train.shape[0], x_train.shape[1]*x_train.shape[2]])
print(x_mnist.shape)

# run t-SNE
tsne = TSNE(n_components=2, verbose=0)
%time z = tsne.fit_transform(x_mnist)

# load to dataframe
df = pd.DataFrame()
df["y"] = y_train
df["comp-1"] = z[:,0]
df["comp-2"] = z[:,1]

# plot it
sns.scatterplot(x="comp-1", y="comp-2", hue=df.y.tolist(),
                palette=sns.color_palette("hls", 10),
                data=df).set(title="Limited MNIST data t-SNE projection")
```
